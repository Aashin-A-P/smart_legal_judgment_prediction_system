```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated 20-epoch GAN training data
epochs = np.arange(1, 21)

# Typical GAN loss patterns
g_loss = np.clip(np.linspace(2.5, 0.9, 20) + np.random.normal(0, 0.05, 20), 0.8, None)
d_loss = np.clip(np.linspace(1.8, 0.95, 20) + np.sin(np.linspace(0, 3*np.pi, 20))*0.1, 0.9, None)
fid = np.linspace(120, 45, 20) + np.random.normal(0, 2, 20)  # lower is better

fig, ax1 = plt.subplots(figsize=(10,6))
ax2 = ax1.twinx()

# Loss curves
ax1.plot(epochs, g_loss, 'o-', color='#4c72b0', linewidth=2, label='Generator Loss')
ax1.plot(epochs, d_loss, 's--', color='#c44e52', linewidth=2, label='Discriminator Loss')

# FID on secondary axis
ax2.plot(epochs, fid, '^-', color='#55a868', linewidth=2.5, label='FID Score')

# Aesthetics
ax1.set_xlabel("Epochs", fontsize=12)
ax1.set_ylabel("Loss", color='#4c72b0', fontsize=12)
ax2.set_ylabel("FID (↓ Better)", color='#55a868', fontsize=12)

ax1.set_title("⚖️ GAN Training Dynamics: Loss & FID Over Epochs", fontsize=14, fontweight='bold')
ax1.grid(True, linestyle='--', alpha=0.6)

# Legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right', fontsize=10, frameon=True)

plt.tight_layout()
plt.show()
```
