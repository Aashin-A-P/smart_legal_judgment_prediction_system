```python
c = []
for i in range(df1.shape[0]):
    c.extend(eval(df1.loc[i, "charges"]))
```

```python
dc = Counter(c)
```

```python
dc
```

```python
cf = pd.DataFrame(list(dc.items()), columns=["charges", "count"])

# Take top 10 by count
top10 = cf.sort_values(by="count", ascending=False).head(10)

# Plot with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="count", y="charges", data=top10, palette="viridis")

plt.title("Top 10 Statutes by Count")
plt.xlabel("Count")
plt.ylabel("Charges")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top10["count"], labels=top10["charges"], autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Top 10 Charges Distribution")
plt.show()
```
