```python
# ================================
# Majority Baseline Classifier
# ================================

import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("batch_1.csv")

X = df["text"]   # not used here, since majority baseline ignores features
y = df["label"]

# Find majority class
majority_class = y.value_counts().idxmax()
print("Majority class:", majority_class)

# Predict everything as majority class
y_pred = [majority_class] * len(y)

# Accuracy
acc = accuracy_score(y, y_pred)
print("Majority Baseline Accuracy:", acc)

# Classification report
print("\nClassification Report:")
print(classification_report(y, y_pred, target_names=["Rejected (0)", "Accepted (1)", "Partially Accepted (2)"]))

# Confusion Matrix
cm = confusion_matrix(y, y_pred)

plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Rejected (0)", "Accepted (1)", "Partially Accepted (2)"],
            yticklabels=["Rejected (0)", "Accepted (1)", "Partially Accepted (2)"])
plt.title("Majority Baseline - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```
